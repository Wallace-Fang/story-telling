```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
video_games_df = pd.read_csv("Video_Games.csv")
```

```python
percent_missing = video_games_df.isnull().sum() * 100 / len(video_games_df)
print(percent_missing)
```

```python
video_games_df.dropna(subset=['Name', 'Platform', 'Year_of_Release', 'Genre', 'Publisher'], inplace=True)
```

```python
percent_missing = video_games_df.isnull().sum() * 100 / len(video_games_df)
print(percent_missing)
```

```python
video_games_df.shape
```

```python
video_games_platforms = video_games_df['Platform'].value_counts()
platform_percentage = video_games_platforms / len(video_games_df) * 100
print(platform_percentage)
```

```python
platform_percentage.plot(kind='bar')
plt.ylabel('Percentage')
plt.show()
```

Given the distribution of the bar graph it looks like the Sony PS2 is the most popular gaming platform with 12.96% and Nintendo DS trailing closely behind at 12.93%. Interestingly despite the data set supposely being updated annually (last updated 2 days ago at the time of exploring this dataset), newer consoles such as PS5, Nintendo Switch, and Xbox Series X are not seen in this graph nor the distribution shown above it. Lets look at the latest oldest and newest data for the video games to see if the dataset is actually up to date. 

```python
print(video_games_df['Year_of_Release'].max())
```

```python
video_games_df['Year_of_Release'] = video_games_df['Year_of_Release'].astype(int)
```

```python
video_games_years = video_games_df['Year_of_Release'].value_counts()
years_percentage = video_games_years / len(video_games_df) * 100
```

```python
years_percentage.plot(kind='bar')
plt.ylabel('Percentage')
plt.show()
```

```python
total_sales_by_year = video_games_df.groupby(video_games_df['Year_of_Release'])['Global_Sales'].sum()
```

```python
total_sales_by_year.plot()

plt.xlabel("Years")
plt.ylabel("Total Global Video Game Sales (Millions)")
plt.title("Global Video Games Sales over the Years")
plt.show()
```

Video games saw a healthy growth overtime since the start in 1980 to 2010 then it rapidly declined. Why is that? Perhaps its just because video games became less popular and sales were at its worst in 2017 to 2020 . Although video game popularity should have increased in 2020 because of the pandamic when everyone was indoors during the lockdown. The only conclusion that could be made is that the dataset does not accurate represent video games after 2010. Lets investigate the distribution of sales by region, I would expect to have the highest share of sales. 

```python
total_NA = video_games_df['NA_Sales'].sum()/video_games_df['Global_Sales'].sum()
total_EU = video_games_df['EU_Sales'].sum()/video_games_df['Global_Sales'].sum()
total_JP = video_games_df['JP_Sales'].sum()/video_games_df['Global_Sales'].sum()
total_others = video_games_df['Other_Sales'].sum()/video_games_df['Global_Sales'].sum()
```

```python
y = np.array([total_NA, total_EU, total_JP, total_others])
mylabels=['NA_Sales','EU_Sales','JP_Sales','Other_Sales' ]
print(y)
plt.pie(y, labels = mylabels)
plt.title("Regional Share of Video Game Sales")
plt.show() 
```

Looks like NA sales makes up just under half of all video games sales with 49.2% while EU is a little over a quarter at 27.2 and Japan's sale making up 14.6% of total global sales. Although this is good for historical data, many of these consoles and platforms do not exist anymore. Perhaps we should narrow our scope to more recent years to predict the current landscape of video games. According to research, we are currently on the ninth generation of video games consoles; however, this generation started in 2020. As we can see the last release date we have is 2020 which is when the lastest consoles were released. Maybe that is the correlation between the lack of data? Anyways instead of focusing on the ninth gen, lets go back one generation to the eigth-gen which started in 2012. Lets use the data to predict from that era to predict the sales for the current generation. 

```python
eigth_gen_df = video_games_df[video_games_df['Year_of_Release'] >=2012]
eigth_gen_df=eigth_gen_df.sort_values(by=['Year_of_Release'])
```

```python
eigth_gen_df.shape
```

After subsetting the original dataframe to only include data from the year 2012 or later, we're down to 2888 rows of data, significantly less than the original 16416. Lets look at the break down of the years of release in this new dataframe.

```python
eigth_gen_years = eigth_gen_df['Year_of_Release'].value_counts()
eigth_gen_years_percentage = eigth_gen_years  / len(eigth_gen_df) * 100
print(eigth_gen_years_percentage)
```

```python
eigth_gen_years_percentage.plot(kind='bar', color=['red', 'blue', 'green', 'grey', 'purple'])
plt.xlabel('Years')
plt.ylabel('Percentages')
plt.show()
```

```python
eigth_gen_platforms = eigth_gen_df['Platform'].value_counts()
eigth_gen_platform_percentage = eigth_gen_platforms / len(eigth_gen_df) * 100
print(eigth_gen_platform_percentage)
```

```python
eigth_gen_platform_percentage
```

Looks like at the year 2012, companies were still releasing games on older platforms, lets keep them since users will most likely not immediately transition to newer consoles immediately. Although we will only look at the game sales in relation to their direct competitor.

```python
x = ["Playstation", "XBox", "Nintendo"]
last_gen = [eigth_gen_platform_percentage['PS3'], eigth_gen_platform_percentage['X360'], eigth_gen_platform_percentage['DS']]
curr_gen = [eigth_gen_platform_percentage['PS4'], eigth_gen_platform_percentage['XOne'], eigth_gen_platform_percentage['3DS']]
pc = eigth_gen_platform_percentage['PC']
```

```python
plt.bar(x,last_gen, color='gray', label="last_gen")
plt.bar(x, curr_gen, bottom=last_gen, color='b', label="curr_gen")
plt.bar("PC", pc, color='g')
plt.xlabel("Platforms")
plt.ylabel("% of Video Games Using this Platform")
plt.title("Share of Consoles/Platforms")

plt.show()
```

We can see that Playstation is by far the most popular platform/console for consumers with the PS3 and PS4 being the most popular console out of the 4 major ones. The share of PS3/PS4 almost made up the combined share of the XBox 360 and XBox One combined. Although the nintendo 3DS shares are more than the PS4s indiciation the 3DS is the most popular current gen platform.
